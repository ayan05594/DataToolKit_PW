
1. What is the difference between a function and a method in Python?
Ans) 
Function	Method
A standalone block of code that can be called from anywhere in the code. It is not associated with any object.	A function that is associated with an object and is called on an instance of a class. It can access and modify the object's attributes.
	
Example:
def greet(name):  # Function
    return f"Hello, {name}!"

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # Method
        return f"Hello, {self.name}!"
print(greet("Alice"))  # Calls the function
person = Person("Bob")
print(person.greet())  # Calls the method


2. Explain the concept of function arguments and parameters in Python.
Ans)
Parameters: Variables defined in the function signature that accept values when the function is called.
Arguments: The actual values passed to the function when it is called.

6. Explain the concept of generators in Python and how they are defined.
Ans) Generators are a special type of iterator that yield values one at a time and maintain their state between calls. They are defined using functions with the yield statement.

def countdown(n):
    while n > 0:
        yield n
        n -= 1

for number in countdown(5):
    print(number)  # Outputs: 5, 4, 3, 2, 1


7. What are the advantages of using generators over regular functions?
Ans) Memory Efficiency: Generators produce items one at a time, thus using less memory than storing an entire list.
Lazy Evaluation: They only compute values as needed, which can lead to performance improvements in large datasets.
def generate_numbers(n):
    for i in range(n):
        yield i * i

# This will only compute values as needed
for num in generate_numbers(10):
    print(num)  # Outputs squares of numbers from 0 to 9

10. What is the difference between `map()`, `reduce()`, and `filter() functions in Python?
Ans) map(): Applies a function to every item in an iterable and returns an iterator of results.
reduce(): Applies a rolling computation to sequential pairs of values in an iterable (requires importing functools).
filter(): Filters items from an iterable based on a function that returns True or False.

from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map()
squared = list(map(lambda x: x ** 2, numbers))  # [1, 4, 9, 16, 25]

# reduce()
summed = reduce(lambda x, y: x + y, numbers)  # 15

# filter()
evens = list(filter(lambda x: x % 2 == 0, numbers))  # [2, 4]

print("Squared:", squared)
print("Summed:", summed)
print("Evens:", evens)

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    print(f"{number} is {'Prime' if is_prime(number) else 'Not Prime'}")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.terms:
            raise StopIteration
        if self.current == 0:
            self.current += 1
            return self.a
        elif self.current == 1:
            self.current += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.current += 1
            return self.b

# Example usage
fib_iterator = FibonacciIterator(10)
for num in fib_iterator:
    print(num, end=" ")
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage
for power in powers_of_two(5):
    print(power, end=" ")
```

```python
#7.Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()


for line in read_file_line_by_line("example.txt"):
   print(line)
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple
tuples_list = [(1, 3), (4, 1), (2, 2), (5, 0)]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print("Sorted list:", sorted_list)
```

```python
#9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temperatures = [0, 20, 30, 40, 100]
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda x: x not in vowels, s))

input_string = "Hello World"
result = remove_vowels(input_string)
print("String without vowels:", result)
```
