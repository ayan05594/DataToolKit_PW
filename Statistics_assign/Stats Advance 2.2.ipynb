Qns 1) What is hypothesis testing in statistics?
Ans)
Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It involves:
Formulating two competing hypotheses: the null hypothesis (H₀) and the alternative hypothesis (H₁).
Using sample data to decide whether to reject the null hypothesis in favor of the alternative hypothesis.

Qns 2) What is the null hypothesis, and how does it differ from the alternative hypothesis?
Ans) Null Hypothesis (H₀): A statement of no effect, no difference, or no relationship. It assumes that any observed differences are due to random chance.
Alternative Hypothesis (H₁): A statement that contradicts the null hypothesis, suggesting an effect, difference, or relationship exists.

Qns 3) What is the significance level in hypothesis testing, and why is it important?
Ans) The significance level (α) is the probability of rejecting the null hypothesis when it is true (Type I error). Common values are 0.05 or 0.01. It determines the threshold for deciding whether the results are statistically significant.

Qns 4) What does a P-value represent in hypothesis testing?
Ans)The P-value is the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming the null hypothesis is true. It quantifies the strength of evidence against the null hypothesis.

Qns 5) How do you interpret the P-value in hypothesis testing?
Ans) P-value ≤ α: Reject the null hypothesis; the result is statistically significant.
P-value > α: Fail to reject the null hypothesis; there is insufficient evidence to support the alternative hypothesis.

Qns 6) What are Type 1 and Type 2 errors in hypothesis testing?
Ans) Type I Error: Rejecting the null hypothesis when it is true (false positive).
Type II Error: Failing to reject the null hypothesis when it is false (false negative).

Qns 7) What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
Ans) One-tailed test: Tests for an effect in one specific direction (e.g., greater than or less than).
Two-tailed test: Tests for an effect in either direction (e.g., different from).

Qns 8) What is the Z-test, and when is it used in hypothesis testing?
Ans) A Z-test is used to test hypotheses about population means or proportions when:
The sample size is large (n > 30).
The population variance is known.

Qns 9) How do you calculate the Z-score, and what does it represent in hypothesis testing?
Ans) The Z-score represents the number of standard deviations a data point is from the population mean. It is calculated as:
Z = (X - μ) / (σ / √n)
Where:
X: Sample mean
μ: Population mean
σ: Standard deviation
n: Sample size

Qns 10) What is the T-distribution, and when should it be used instead of the normal distribution?
Ans) The T-distribution is used when the sample size is small (n ≤ 30) and the population variance is unknown. It is wider and more spread out than the normal distribution.

Qns 11) What is the difference between a Z-test and a T-test?
Ans) Z-test: Used for large samples and when the population variance is known.
T-test: Used for small samples or when the population variance is unknown.

Qns 12) What is the T-test, and how is it used in hypothesis testing?
Ans) The T-test is used to compare means. It has different types:
One-sample T-test: Compares a sample mean to a population mean.
Independent T-test: Compares means of two independent groups.
Paired T-test: Compares means of paired observations.

Qns 13) What is the relationship between Z-test and T-test in hypothesis testing?
Ans) Both tests compare sample means to population means or other sample means. The choice depends on sample size and whether the population variance is known.

Qns 14) What is a confidence interval, and how is it used to interpret statistical results?
Ans) A confidence interval (CI) is a range of values that is likely to contain the population parameter with a specified probability (e.g., 95%). It provides an estimate of the parameter's precision.

Qns 15) What is the margin of error, and how does it affect the confidence interval?
Ans) The margin of error represents the maximum expected difference between the observed sample statistic and the true population parameter. Larger margins result in wider confidence intervals.

Qns 16) How is Bayes' Theorem used in statistics, and what is its significance?
Ans) Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It is widely used in Bayesian statistics, where prior probabilities are updated with observed data.

Qns 17) What is the Chi-square distribution, and when is it used?
Ans) The Chi-square distribution is used for categorical data and is applied in tests like the Chi-square goodness-of-fit test and Chi-square test of independence.

Qns 18) What is the Chi-square goodness of fit test, and how is it applied?
Ans)The Chi-square goodness-of-fit test assesses whether observed categorical data matches an expected distribution. It compares observed and expected frequencies.

Qns 19) What is the F-distribution, and when is it used in hypothesis testing?
Ans) The F-distribution is used in tests comparing variances (e.g., F-test) or in analysis of variance (ANOVA). It is asymmetrical and depends on two degrees of freedom.

Qns 20) What is an ANOVA test, and what are its assumptions?
Ans) ANOVA (Analysis of Variance) compares means of three or more groups. 
Assumptions:
mality of data within groups.
Homogeneity of variances.
Independence of observations.

Qns 21) What are the different types of ANOVA tests?
Ans)One-way ANOVA: Compares means of groups based on one factor.
Two-way ANOVA: Compares means based on two factors.
Repeated measures ANOVA: Compares means within the same subjects over time.

Qns 22) What is the F-test, and how does it relate to hypothesis testing?
Ans) The F-test is used to compare variances or test the overall significance in ANOVA. It computes the ratio of variances to determine whether group means are significantly different.

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
from scipy.stats import norm
import numpy as np

sample_data = [56, 60, 61, 55, 57, 59, 58, 62]
population_mean = 50
population_std = 5

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
def visualize_type_errors(mu0, mu1, sigma, alpha, sample_size):
    x = np.linspace(mu0 - 4 * sigma, mu1 + 4 * sigma, 1000)
    null_dist = norm.pdf(x, mu0, sigma / np.sqrt(sample_size))
    alt_dist = norm.pdf(x, mu1, sigma / np.sqrt(sample_size))
    critical_value = mu0 + norm.ppf(1 - alpha) * sigma / np.sqrt(sample_size)

    plt.plot(x, null_dist, label='Null Distribution')
    plt.plot(x, alt_dist, label='Alternative Distribution')
    plt.axvline(critical_value, color='red', linestyle='--', label='Critical Value')
    plt.fill_between(x, null_dist, where=(x > critical_value), color='red', alpha=0.2, label='Type I Error')
    plt.fill_between(x, alt_dist, where=(x <= critical_value), color='blue', alpha=0.2, label='Type II Error')
    plt.legend()
    plt.show()

visualize_type_errors(mu0=50, mu1=55, sigma=10, alpha=0.05, sample_size=30)
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    t_critical = t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

data = [54, 56, 58, 60, 62, 64, 66]
confidence_level = 0.95
ci = confidence_interval(data, confidence=confidence_level)

print(f"Confidence Interval ({confidence_level*100}%): {ci[0]:.2f} to {ci[1]:.2f}")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    t_critical = t.ppf((1 + confidence) / 2, df=sample_size - 1)
    moe = t_critical * (sample_std / np.sqrt(sample_size))
    return moe

data = [54, 56, 58, 60, 62, 64, 66]
confidence_level = 0.95
moe = margin_of_error(data, confidence=confidence_level)

print(f"Margin of Error at {confidence_level*100}% confidence level: {moe:.2f}")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
import numpy as np

def calculate_expected_frequencies(observed_data):
    row_totals = np.sum(observed_data, axis=1)
    col_totals = np.sum(observed_data, axis=0)
    grand_total = np.sum(observed_data)

    expected_frequencies = np.outer(row_totals, col_totals) / grand_total
    return expected_frequencies

observed_data = np.array([[50, 30],
                           [20, 40]])

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Data:\n", observed_data)
print("\nExpected Frequencies:\n", expected_frequencies)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit(observed, expected):
    chi2_stat, p_value = chisquare(observed, expected)
    return chi2_stat, p_value

observed_data = np.array([50, 30, 20])
expected_data = np.array([40, 40, 20])

chi2_stat, p_value = goodness_of_fit(observed_data, expected_data)

print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data fits the expected distribution.")
```

```python
#15. Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

def plot_chi_square_distribution(degrees_of_freedom, num_samples=10000):
    samples = np.random.chisquare(degrees_of_freedom, num_samples)
    
    plt.figure(figsize=(8, 6))
    plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', label=f'Chi-square df={degrees_of_freedom}')
    
    x = np.linspace(0, np.max(samples), 100)
    y = chi2.pdf(x, degrees_of_freedom)
    plt.plot(x, y, 'r-', label='Theoretical Distribution')
    
    plt.title(f'Chi-square Distribution (df={degrees_of_freedom})')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

degrees_of_freedom = 4
plot_chi_square_distribution(degrees_of_freedom)
```

```python
#16.  Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
    
    p_value = 1 - f.cdf(f_statistic, df1, df2)
    
    return f_statistic, p_value

sample1 = np.array([25, 30, 35, 40, 45])
sample2 = np.array([20, 22, 28, 32, 38])

f_statistic, p_value = f_test(sample1, sample2)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances are not significantly different.")
```

```python
#17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results
import numpy as np
import scipy.stats as stats

def anova_test(*groups):
    f_statistic, p_value = stats.f_oneway(*groups)
    return f_statistic, p_value

# Example data for three groups
group1 = np.array([23, 25, 28, 30, 32])
group2 = np.array([22, 24, 27, 29, 33])
group3 = np.array([21, 26, 28, 31, 35])

f_statistic, p_value = anova_test(group1, group2, group3)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: At least one group mean is different.")
else:
    print("\nFail to reject the null hypothesis: The group means are not significantly different.")
```

```python
#18.  Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

np.random.seed(42)
group1 = np.random.normal(loc=5, scale=1, size=30)
group2 = np.random.normal(loc=6, scale=1, size=30)
group3 = np.random.normal(loc=7, scale=1, size=30)

data = pd.DataFrame({
    'Group': ['Group 1'] * len(group1) + ['Group 2'] * len(group2) + ['Group 3'] * len(group3),
    'Value': np.concatenate([group1, group2, group3])
})

f_statistic, p_value = stats.f_oneway(group1, group2, group3)

print(f"F-statistic: {f_statistic:.2f}, p-value: {p_value:.4f}")

plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', data=data)
plt.title('Boxplot of Groups')
plt.xlabel('Group')
plt.ylabel('Value')
plt.grid(True)
plt.show()

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the group means.")
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, group_col, value_col):
    groups = data.groupby(group_col)[value_col].apply(list)
    
    normality_results = {}
    for group in groups.index:
        stat, p_value = stats.shapiro(groups[group])
        normality_results[group] = p_value > 0.05
    
    independence_result = "Assumed based on study design"
    
    stat, p_value = stats.levene(*groups)
    equal_variance_result = p_value > 0.05
    
    print("Normality Check Results:")
    for group, is_normal in normality_results.items():
        print(f"{group}: {'Normal' if is_normal else 'Not Normal'}")
    
    print("\nIndependence Check Result:")
    print(independence_result)
    
    print("\nEqual Variance Check Result:")
    print(f"Equal Variance: {'Yes' if equal_variance_result else 'No'}")
    
    plt.figure(figsize=(12, 6))
    for i, group in enumerate(groups.index):
        plt.subplot(1, len(groups), i + 1)
        stats.probplot(groups[group], dist="norm", plot=plt)
        plt.title(f'Q-Q Plot for {group}')
    plt.tight_layout()
    plt.show()

np.random.seed(42)
group1 = np.random.normal(loc=5, scale=1, size=30)
group2 = np.random.normal(loc=6, scale=1, size=30)
group3 = np.random.normal(loc=7, scale=1, size=30)

data = pd.DataFrame({
    'Group': ['Group 1'] * len(group1) + ['Group 2'] * len(group2) + ['Group 3'] * len(group3),
    'Value': np.concatenate([group1, group2, group3])
})

check_anova_assumptions(data, 'Group', 'Value')
```

```python
#20.  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
factor_a = ['A1', 'A1', 'A1', 'A1', 'A2', 'A2', 'A2', 'A2']
factor_b = ['B1', 'B1', 'B2', 'B2', 'B1', 'B1', 'B2', 'B2']
values = [5, 6, 7, 8, 6, 7, 8, 9]

data = pd.DataFrame({
    'Factor_A': factor_a,
    'Factor_B': factor_b,
    'Value': values
})

model = ols('Value ~ C(Factor_A) * C(Factor_B)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)

plt.figure(figsize=(10, 6))
sns.pointplot(x='Factor_A', y='Value', hue='Factor_B', data=data, 
               markers=['o', 's'], linestyles=['-', '--'], errorbar=None)
plt.title('Interaction Plot of Factor A and Factor B')
plt.xlabel('Factor A')
plt.ylabel('Mean Value')
plt.grid(True)
plt.show()
```

```python
#22.  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
group1 = np.random.normal(loc=5, scale=1, size=30)
group2 = np.random.normal(loc=6, scale=1, size=30)
group3 = np.random.normal(loc=7, scale=1, size=30)

data = pd.DataFrame({
    'Group': ['Group 1'] * len(group1) + ['Group 2'] * len(group2) + ['Group 3'] * len(group3),
    'Value': np.concatenate([group1, group2, group3])
})

f_statistic, p_value = stats.f_oneway(group1, group2, group3)

print(f"F-statistic: {f_statistic:.2f}, p-value: {p_value:.4f}")

plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', data=data)
plt.title('Boxplot of Groups')
plt.xlabel('Group')
plt.ylabel('Value')
plt.grid(True)
plt.show()

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the group means.")
```

```python
#24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample_data = np.random.normal(loc=10, scale=2, size=30)  
sigma_squared_0 = 4 
n = len(sample_data)
sample_variance = np.var(sample_data, ddof=1)  
chi_square_statistic = (n - 1) * sample_variance / sigma_squared_0

alpha = 0.05  
df = n - 1  
critical_value_lower = stats.chi2.ppf(alpha / 2, df)
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)

print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"Critical Value Lower: {critical_value_lower:.4f}")
print(f"Critical Value Upper: {critical_value_upper:.4f}")

if chi_square_statistic < critical_value_lower or chi_square_statistic > critical_value_upper:
    print("Reject the null hypothesis: There is a significant difference in population variance.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in population variance.")
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

successes1 = 30  
n1 = 100         

successes2 = 50  
n2 = 120         

p1 = successes1 / n1
p2 = successes2 / n2

p_pool = (successes1 + successes2) / (n1 + n2)

z_statistic = (p1 - p2) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

alpha = 0.05  
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the proportions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the proportions.")
```

```python
#26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
group1 = np.random.normal(loc=10, scale=2, size=30)
group2 = np.random.normal(loc=12, scale=3, size=30)

var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

f_statistic = var1 / var2

df1 = len(group1) - 1
df2 = len(group2) - 1
p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

alpha = 0.05
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the variances.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the variances.")

plt.figure(figsize=(10, 6))
sns.boxplot(data=[group1, group2], palette="Set2")
plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.title('Boxplot of Group Variances')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
#27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
observed_counts = np.random.choice(categories, size=1000, p=[0.1, 0.3, 0.4, 0.2])
observed_freq = pd.Series(observed_counts).value_counts().sort_index()

expected_freq = pd.Series([100, 300, 400, 200], index=categories)

chi2_statistic, p_value = stats.chisquare(observed_freq, f_exp=expected_freq)

alpha = 0.05
print(f"Chi-square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution fits the expected distribution.")

plt.figure(figsize=(10, 6))
plt.bar(observed_freq.index, observed_freq.values, alpha=0.6, label='Observed', color='blue')
plt.bar(expected_freq.index, expected_freq.values, alpha=0.6, label='Expected', color='orange')
plt.title('Observed vs Expected Frequencies')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
