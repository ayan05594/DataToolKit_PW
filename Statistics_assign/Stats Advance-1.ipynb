Define the z-statistic and explain its relationship to the standard normal distribution. How is the z-statistic used in hypothesis testing?
Ans) The z-statistic is a measure that describes how many standard deviations a data point or sample mean is away from the population mean. It is calculated using the formula:
z=(X−μ)/σ
Where,
X is the observed data point or sample mean
μ  is the population mean
σ is the standard deviation of the population

Relationship to the Standard Normal Distribution
The z-statistic is based on the standard normal distribution, which is a normal distribution with:
A mean of 0 (μ=0)
A standard deviation of 1 (σ=1)
The z-score represents the position of a value on the standard normal distribution. For example:
A z-score of 0 means the value is exactly at the mean.
A z-score of +1 or -1 means the value is 1 standard deviation above or below the mean, respectively.

Use in Hypothesis Testing
The z-statistic plays a crucial role in hypothesis testing, particularly for large sample sizes (n>=30) or when the population standard deviation is known. It is used as follows:
1.	State the Hypotheses:
o	Null Hypothesis (H0): No effect or no difference (μ=μ0).
o	Alternative Hypothesis (Ha): Indicates an effect or difference (μ!=μ0, μ>μ0 , μ<μ0).
2.	Calculate the Z-Statistic:
Compute the z-score using the sample data.
3.	Determine the Critical Value(s):
From the standard normal distribution table, find the critical z-value(s) corresponding to the chosen significance level.
4.	Compare Z-Statistic with Critical Value(s):
o	If the z-statistic falls in the rejection region (beyond the critical z-value), reject H0
o	If it does not, fail to reject H0
5.	P-Value Approach:
Alternatively, compare the calculated p-value (area under the curve beyond the z-statistic) to the significance level α. Reject H0 if p<=α.

What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is very small (e.g., 0.01)?
Ans) The p-value is the probability of obtaining a test statistic as extreme as, or more extreme than, the observed value, assuming the null hypothesis (H0H_0H0) is true. It quantifies the strength of the evidence against the null hypothesis.
p-value=P(Test Statistic>=Observed Value | H0 is true).
In hypothesis testing, the p-value is used to decide whether to reject or fail to reject the null hypothesis. The process involves:
Set a significance level α: Common choices for α\alphaα are 0.05, 0.01, or 0.10, representing the threshold for determining statistical significance.
Compare the p-value to α:
o	If p<=α : Reject the null hypothesis (H0). The data provides sufficient evidence to support the alternative hypothesis (Ha).
o	If p>α: Fail to reject the null hypothesis (H0). The data does not provide strong enough evidence against H0.

If the p-value is very small, such as 0.01, it indicates that:
1.	Strong Evidence Against H0: The observed result is highly unlikely under the assumption that H0
2.	Statistical Significance: At a significance level of α=0.05, a p-value of 0.01 would lead to rejecting H0.
3.	Practical Implication: There is a strong indication that the effect, difference, or relationship observed in the data is real and not due to random chance

Question 4:
Under what conditions is the binomial distribution used, and how does it relate to the Bernoulli distribution?
Ans) Conditions for Using the Binomial Distribution:
1.	Fixed Number of Trials: The experiment is repeated a fixed number of times, denoted by n.
2.	Binary Outcomes: Each trial has only two possible outcomes, typically referred to as "success" and "failure."
3.	Independent Trials: The outcome of one trial does not affect the outcome of any other trial.
4.	Constant Probability of Success: The probability of success, denoted by p, remains constant across all trials.
Relation to the Bernoulli Distribution:
The binomial distribution can be thought of as a generalization of the Bernoulli distribution:
•	A Bernoulli distribution models a single trial with two outcomes: success (with probability p) and failure (with probability 1−p).
•	The binomial distribution models the total number of successes in n independent Bernoulli trials.
the Bernoulli distribution is a special case of the binomial distribution with n=1.


Question 5:  
What are the key properties of the Poisson distribution, and when is it appropriate to use this distribution? 
Ans) Key Properties of the Poisson Distribution:
1.	Discrete Distribution: The Poisson distribution models the number of occurrences of an event in a fixed interval of time, space, or other dimensions.
2.	Non-Negative Integer Values: The possible values of the random variable X (representing the number of events) are 0,1,2,…
3.	Parameter (λ): The distribution is defined by a single parameter λ, which represents the average number of events in the given interval. It must be positive (λ>0).
4.	Mean and Variance:
o	Mean: E(X)=λ
o	Variance: Var(X)=λ
5.	Memory lessness: While not strictly memoryless like the exponential distribution, the Poisson process (on which the distribution is based) assumes events occur independently in disjoint intervals.
When to Use the Poisson Distribution:
The Poisson distribution is appropriate when the following conditions are met:
1.	Rare Events: The distribution is suitable for modeling the occurrence of rare events.
2.	Independent Events: The occurrences of events are independent of each other.
3.	Constant Rate (λ): The average rate of occurrence (λ) is constant over the interval.
4.	Non-Overlapping Intervals: Events in disjoint intervals do not influence one another.
5.	Small Probability per Subinterval: In a sufficiently small subinterval, the probability of more than one event is negligible.

Explain the Central Limit Theorem (CLT) with example.
Ans) The Central Limit Theorem (CLT) is a fundamental statistical principle that states:
When independent random variables are added, their normalized sum tends to follow a normal distribution (Gaussian distribution), regardless of the original distribution of the variables, provided the sample size is sufficiently large.
Conditions for CLT:
1.	Independence: The data points should be independent of each other.
2.	Sample Size: A larger sample size (n>=30) ensures better approximation to normality.
3.	Finite Variance: The population must have a finite variance.
Example:

```python
#Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample
#size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to
#reject the null hypothesis?

# Task: Write Python code to calculate the z-score and p-value for the given data.

#Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.
import scipy.stats as stats

# Given data
sample_mean = 105
population_mean = 100
standard_deviation = 15
sample_size = 25
alpha = 0.05  # significance level

# Calculate the z-score
standard_error = standard_deviation / (sample_size ** 0.5)
z_score = (sample_mean - population_mean) / standard_error

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Decision based on p-value
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

# Output results
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Decision: {decision}")
```

```python
#Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.
#Generate 1,000 samples and plot the distribution. What is the expected mean and variance?

#Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.

#Objective: Understand the properties of a binomial distribution and verify them through simulation
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n_trials = 10  # Number of trials
p_success = 0.6  # Probability of success
n_samples = 1000  # Number of samples to generate

# Simulate the binomial distribution
data = np.random.binomial(n=n_trials, p=p_success, size=n_samples)

# Calculate the expected mean and variance
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)

# Calculate the sample mean and variance
sample_mean = np.mean(data)
sample_variance = np.var(data)

# Plot the distribution
plt.hist(data, bins=range(n_trials + 2), density=True, alpha=0.75, color='blue', edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n_trials + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print the results
print(f"Expected Mean: {expected_mean:.2f}")
print(f"Expected Variance: {expected_variance:.2f}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
```
