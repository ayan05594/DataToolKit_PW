3. Explain the following termsg1)M The difference in np.array, np.asarray and np.asanyarrayX2) M The difference between Deep copy and shallow copyX
In NumPy, the functions np.array, np.asarray, and np.asanyarray are used to convert input data into arrays, but they differ in their behavior. np.array creates a new array by default, always copying the data, even if the input is already a NumPy array. This ensures the creation of a completely independent array in memory. On the other hand, np.asarray converts an input object into an array but avoids copying the data if the input is already a NumPy array, making it more memory-efficient. np.asanyarray is similar to np.asarray, but it preserves the type of the input if it is a subclass of an ndarray, such as a masked_array. While np.asarray converts any input to a standard ndarray, np.asanyarray retains the subclass type if one exists.

When considering copying in Python, shallow and deep copies are essential concepts. A shallow copy creates a new object but does not create independent copies of the elements within it. Instead, it copies references to the original objects, meaning that changes to mutable elements (like lists or arrays) in the shallow copy also affect the original object. Conversely, a deep copy creates a completely independent duplicate of the object, including all nested elements. This ensures that changes made to the deep copy do not affect the original object and vice versa. Shallow copies are typically created using copy.copy(), whereas deep copies require copy.deepcopy() for full independence.

```python
#15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:a) Plot a histogram of the data with 30 bins. 
#b) Overlay a line plot representing the normal distribution's probability density function (PDF).
#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
#d) Set the title of the plot as 'Histogram with PDF Overlay'.

data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='gray')

x = np.linspace(min(data), max(data), 1000)
pdf = norm.pdf(x, loc=0, scale=1)

plt.plot(x, pdf, color='red', linewidth=2, label='PDF')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

```python
#17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot

x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)

quadrants = np.where((x >= 0) & (y >= 0), 'Quadrant I',
            np.where((x < 0) & (y >= 0), 'Quadrant II',
            np.where((x < 0) & (y < 0), 'Quadrant III', 'Quadrant IV')))

data = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='tab10')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.axvline(0, color='black', linestyle='--', linewidth=0.8)
plt.legend(title='Quadrants')
plt.show()
```

```python
#18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'8

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y', width=800, height=400)
p.line(x, y, legend_label="Sine", line_width=2)

p.grid.grid_line_color = 'gray'
show(p)
```

```python
#19.  Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'8

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 10, size=5)

source = ColumnDataSource(data=dict(categories=categories, values=values))

p = figure(x_range=categories, title="Random Categorical Bar Chart", 
           x_axis_label='Categories', y_axis_label='Values', 
           width=800, height=400)

p.vbar(x='categories', top='values', width=0.9, source=source, 
       legend_field="categories", line_color='white', 
       fill_color='blue')

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

p.grid.grid_line_color = 'gray'
p.legend.orientation = 'horizontal'
p.legend.location = 'top_center'

show(p)
```
